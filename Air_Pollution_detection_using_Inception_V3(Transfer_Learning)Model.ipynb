## Transfer Learning Inception V3 using Keras

## **Step-1: Importing Required Libraries**

```python
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession

config = ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.5
config.gpu_options.allow_growth = True
session = InteractiveSession(config=config)
```

```python
# import the libraries as shown below

from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
#from keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
#import matplotlib.pyplot as plt
```

## **Step-2: Assigning Datasets path**

```python
# re-size all the images to this
IMAGE_SIZE = [224, 224]

#train_path = 'D/content/drive/MyDrive/Air_Pollution_Dataset/train'
#valid_path = '/content/drive/MyDrive/Air_Pollution_Dataset/Validate'

train_path = 'C:/Users/onkar/Desktop/LY P/Input_dataset-20230125T202411Z-001/Input_dataset/train'
valid_path = 'C:/Users/onkar/Desktop/LY P/Input_dataset-20230125T202411Z-001/Input_dataset/validate'
```

## **Step-3: Importing Inception v3 Layers and adding preprocessing layer to the input layer**

```python

# Here we will be using imagenet weights

inception = InceptionV3(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)


```

```python
# don't train existing weights
for layer in inception.layers:
    layer.trainable = False
```

```python
  # useful for getting number of output classes
folders = glob('C:/Users/onkar/Desktop/LY P/Input_dataset-20230125T202411Z-001/Input_dataset/train/*')
```

```python
# our layers
x = Flatten()(inception.output)
```

# **Step-4: Creating the Model object**

```python
prediction = Dense(len(folders), activation='softmax')(x)

# create a model object
model = Model(inputs=inception.input, outputs=prediction)
```

```python

# view the structure of the model
model.summary()
```

# **Step-5: Assigning Cost and Optimization method to the model**

```python
# tell the model what cost and optimization method to use
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

# **Step-6: Performing Preprocessing on the input Data**

```python
# Use the Image Data Generator to import the images from the dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python

training_set = train_datagen.flow_from_directory('C:/Users/onkar/Desktop/LY P/Input_dataset-20230125T202411Z-001/Input_dataset/train',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/Air_Pollution_Dataset/Validate',
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

# **Step-7: Fit the Model**

```python


r = model.fit_generator(
  training_set,
  validation_data=test_set,
  epochs=10,
  steps_per_epoch=len(training_set),
  validation_steps=len(test_set)
)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot the loss
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```
